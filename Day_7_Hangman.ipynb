## The HangMan Game

Hangman is a classic word game in which you must guess as many secret words as you can before time runs out! [Read More](https://hangmanwordgame.com/?fca=1&success=0#/)

```python
#Step 1
import random

stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']

word_list = ["aardvark", "baboon", "camel"]
choices = random.choice(word_list)
lives = 6

choices_list =[]
for i in range(0,len(choices)):
  choices_list.append("-")

while '-' in  choices_list and lives > 0:
  print(stages[lives])

  #TODO-2 - Ask the user to guess a letter and assign their answer to a variable called guess. Make guess lowercase.
  user = input("Guess a word: ").lower()

  # Initialize a flag to check if the guess is correct
  guess_correct = False

  #Check if the letter the user guessed (guess) is one of the letters in the chosen_word.
  for index,letter in enumerate(choices):
    if user == letter:
      choices_list[index]=letter
      guess_correct = True

  if not guess_correct:
    lives -=1
    print(f"Incorrect guess. You have {lives} lives left.")

  print("".join(choices_list))

# Check the final state
if lives <= 0:
    print(stages[lives])
    print("You lose.")
else:
    print("Congratulations, you guessed the word!")

```
